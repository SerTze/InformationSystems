## Analytics for GraphX outputs

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# degree centrality
df = pd.read_csv('/home/user/workspace-graphx/outputs/degreeCentrality.txt', sep=" ", header=None)

# degrees is the second row
degrees = df[1]

# find max degree
max_degree = max(degrees)
print(f"\nMax degree: {max_degree}")
```

We can see that the max degree is in accordance with the maximum degree d_max on the [dataset statistics](http://konect.cc/networks/web-Google/)

```python
# get degrees dataframe frequencies
degrees, frequencies = np.unique(degrees, return_counts=True)

# plot the degree distribution log plot
plt.scatter(degrees, frequencies, s=8)
plt.xlabel("Degree")
plt.ylabel("Frequency")
plt.title("Degree Distribution Log Plot")
plt.xscale('log')
plt.yscale('log')

plt.show()
```

We can see that the degree distribution log plot is in accordance with the plot on the [dataset plots](http://konect.cc/plot/degree.a.web-Google.full.png)

```python
# shortest paths
sp = pd.read_csv('/home/user/workspace-graphx/outputs/shortestPaths.txt', sep=" ", header=None)

# shortest paths is the second row
sps = sp[1]

# remove inf values
sps = sps[sps != np.inf]

mean_sp = sps.mean()
print(f"Mean shortest path: {mean_sp}")
```

```python
# weakly connected components
wcc = pd.read_csv('/home/user/workspace-graphx/outputs/weaklyConnectedComponents.txt', sep=" ", header=None)

# weakly connected components is the second row
wcc = wcc[1]

# find instances of most frequent value

print(f"Number of nodes in largest WCC (with index {wcc.value_counts().idxmax()}): {wcc.value_counts().max()}")
```

We can see that the number of nodes in largest WCC is in accordance with the nodes in largest WCC on the [dataset statistics](https://snap.stanford.edu/data/web-Google.html)
